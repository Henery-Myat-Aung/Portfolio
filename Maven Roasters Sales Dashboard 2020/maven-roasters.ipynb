```python
# A new Dataframe without date and time data for aggregation
filtered_df = df.drop(columns = ['transaction_time','transaction_date'])

# calculating total amount from transaction quantity and unit price
new_df = filtered_df.assign(total_amount = filtered_df['transaction_qty'] * filtered_df['unit_price'])

#total sales by store location
new_df.groupby(['store_location']).sum()['total_amount']
```

```python
new_df.describe()
```
