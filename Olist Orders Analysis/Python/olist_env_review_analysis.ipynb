```python
import pandas as pd
from textblob import TextBlob
from unidecode import unidecode
```

```python
#Read Data
df = pd.read_csv('olist_order_reviews_dataset.csv')
```

```python
df
```

```python
df_clean = df.dropna(subset=['review_comment_message']).copy()
```

```python
def clean_text (text):
    if not isinstance(text, str): return '' #if not text return ''
    text = unidecode(text) #change text ascent to unicode
    return text.lower().strip() #lower casae and strip leading and trailing spaces
```

```python
#adding cleaned_review coumn by applying clean text function on review_comment_message column
df_clean['cleaned_review'] = df_clean['review_comment_message'].apply(clean_text)
```

```python
#creating a dictionary to categorize the reviews
def categorize_review(text):
    text = str(text)
    if 'entreg' in text or 'atras' in text or 'chegou' in text or 'correio' in text:
        return 'Delivery / Logistics'
    elif 'deveito' in text or 'quebrad' in text or 'funciona' in text or 'qualidade' in text:
        return 'Product Quality'
    elif 'errad' in text or 'veio' in text or 'diferente' in text:
        return 'Wrong Item'
    else:
        return 'General / Other'
```

```python
df_clean['review_category'] = df_clean['cleaned_review'].apply(categorize_review)
```

```python
df_clean
```

```python
def get_sentiment(text):
    return TextBlob(text).sentiment.polarity
```

```python
df_clean['sentiment_score'] = df_clean['cleaned_review'].apply(get_sentiment)
```

```python
df_clean
```

```python
df_clean['sentiment_label'] = df_clean['sentiment_score'].apply(
lambda x: 'Negative' if x < -0.01 else ('Positive' if x > 0.01 else 'Neutral')
)
```
