## CLean the Data

### Reviews

```python
reviews['review_comment_title'] = reviews['review_comment_title'].fillna('No Title')
```

```python
reviews['review_comment_message'] = reviews['review_comment_message'].fillna('No Comment')
```

<hr>

```python
orders.loc[orders.isnull().any(axis = 1)]
```

#### Orders purchased dates with nulls in approved, delivered dates that are not recent (within 3 months from latest date) present and should be dropped to not skew the data.
<hr>

### Finding latest order date and cutoff date

```python
# changed data type of order_purchase timestamp to datetime to find max date and cutoff date
orders['order_purchase_timestamp'] = pd.to_datetime(orders['order_purchase_timestamp'])
```

```python
orders.dtypes
```

```python
max_date = orders['order_purchase_timestamp'].max()
```

```python
max_date
```

```python
cutoff_date = max_date - pd.DateOffset(months = 3)
```

```python
cutoff_date
```

```python
rows_to_drop = (orders['order_delivered_customer_date'].isnull() & (orders['order_purchase_timestamp'] < cutoff_date))

clean_orders = orders[~rows_to_drop]
```

```python
clean_orders.isnull().sum()
```

#### Orders which are not delivered or those with no records before 3 months from latest date were dropped.
<hr>

### Products

```python
products.isnull().sum()
```

```python
products_NAs = products.loc[products.isnull().any(axis = 1)]
```

```python
products_NAs
```

#### Although product weight and dimensions exist, there were no descriptions. This could indicate faulty records or items not fully cataloged. We need to ensure descriptions are provided by vendors or consider removing such incomplete records if they negatively impact analysis or operations. However for the sake of the project constraints, these products will be dropped for now along with products without dimensions and weight data.

```python
rows_to_drop = (products['product_category_name'].isnull() & products['product_name_lenght'].isnull() &
                products['product_description_lenght'].isnull() & products['product_photos_qty'].isnull())

clean_products = products[~rows_to_drop]
```

```python
clean_products.isnull().sum()
```

```python
# the only product with null product dimensions(height, length, width) is dropped
clean_products = clean_products[~(clean_products['product_weight_g'].isnull())]
```

```python
clean_products.isnull().sum()
```
